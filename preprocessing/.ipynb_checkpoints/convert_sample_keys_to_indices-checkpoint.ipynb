```python
with open(WEM_PKL, 'rb') as f:
    wmap = pickle.load(f)
    
with open(CEM_PKL, 'rb') as f:
    cmap = pickle.load(f)

wem = wmap['embedding_matrix']
w2i = wmap['word2index']
cem = cmap['embedding_matrix']
c2i = cmap['char2index']
```

```python
# 效率有些底下
# def key2index(item):
#     row = item.copy()
#     words1, words2 = row['words1'], row['words2']
#     chars1, chars2 = row['chars1'], row['chars2']
    
#     def word2index(w):
#         return np.array(list(map(lambda x: w2i[x], w.split())))

#     def char2index(c):
#         return np.array(list(map(lambda x: c2i[x], c.split())))
    
#     row['words1'] = word2index(words1)
#     row['words2'] = word2index(words2)
#     row['chars1'] = char2index(chars1)
#     row['chars2'] = char2index(chars2)
    
#     return row

def words2ids(item):
    return np.array(list(map(lambda x: w2i[x],item.split())))

def chars2ids(item):
    return np.array(list(map(lambda x: c2i[x],item.split())))

kn_train_wv_cv_words1, kn_train_wv_cv_words2 = kn_train_wv_cv['words1'], kn_train_wv_cv['words2']
kn_train_wv_cv_chars1, kn_train_wv_cv_chars2 = kn_train_wv_cv['chars1'], kn_train_wv_cv['chars2']
test_wv_cv_words1, test_wv_cv_words2 = test_wv_cv['words1'], test_wv_cv['words2']
test_wv_cv_chars1, test_wv_cv_chars2 = test_wv_cv['chars1'], test_wv_cv['chars2']
kn_train_wids_cids = kn_train_wv_cv.copy()
test_wids_cids = test_wv_cv.copy()
```

```python
kn_train_wids_cids['words1'] = kn_train_wv_cv_words1.map(words2ids)
kn_train_wids_cids['words2'] = kn_train_wv_cv_words2.map(words2ids)
kn_train_wids_cids['chars1'] = kn_train_wv_cv_chars1.map(chars2ids)
kn_train_wids_cids['chars2'] = kn_train_wv_cv_chars2.map(chars2ids)
kn_train_wids_cids
```

```python
test_wids_cids['words1'] = test_wv_cv_words1.map(words2ids)
test_wids_cids['words2'] = test_wv_cv_words2.map(words2ids)
test_wids_cids['chars1'] = test_wv_cv_chars1.map(chars2ids)
test_wids_cids['chars2'] = test_wv_cv_chars2.map(chars2ids)
test_wids_cids
```

```python
KN_TRAIN_WIDS_CIDS_PKL = os.path.join(INTERMEDIATE_DATA_PATH, 'kn_train_wids_cids.pkl')
TEST_WIDS_CIDS_PKL = os.path.join(INTERMEDIATE_DATA_PATH, 'test_wids_cids.pkl')

# kn_train_wids_cids.to_csv(KN_TRAIN_WIDS_CIDS_CSV, sep='\t', index=False)
# test_wids_cids.to_csv(TEST_WIDS_CIDS_CSV, sep='\t', index=False)
with open(KN_TRAIN_WIDS_CIDS_PKL, 'wb') as f:
    pickle.dump(kn_train_wids_cids, f)
```
