```python
# 加载模块
import os
import gc
import pandas as pd
import pickle
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
# 指定数据集文件
DATA_SET_PATH = "/root/mounted/datasets/data_0515/"
KN_CSV = os.path.join(DATA_SET_PATH, "knowledge.csv")
TRAIN_CSV = os.path.join(DATA_SET_PATH, "train.csv")
TEST_CSV = os.path.join(DATA_SET_PATH, "test.csv")
SUBMIT_CSV = os.path.join(DATA_SET_PATH, "submit.csv")
CHAR_EMBED_PKL = os.path.join(DATA_SET_PATH, "char_embed.pkl")
WORD_EMBED_PKL = os.path.join(DATA_SET_PATH, "word_embed.pkl")
QUESTION_PKL = os.path.join(DATA_SET_PATH, "question.pkl")
INTERMEDIATE_DATA_PATH = os.path.join(DATA_SET_PATH, 'intermediate')
KN_TRAIN_CSV = os.path.join(INTERMEDIATE_DATA_PATH, 'kn_train.csv')
KN_TRAIN_WV_CV_CSV = os.path.join(INTERMEDIATE_DATA_PATH, 'kn_train_wv_cv.csv')
TEST_WV_CV_CSV = os.path.join(INTERMEDIATE_DATA_PATH, 'test_wv_cv.csv')
WORDS_NUM = 20891
CHARS_NUM = 3048
WORD_EMBEDDING_DIM = 300
CHAR_EMBEDDING_DIM = 300
```

```python
wv = pickle.load(open(WORD_EMBED_PKL, 'rb'))
len(wv)
```

```python
qs = pickle.load(open(QUESTION_PKL, 'rb'))
type(qs)
```

```python
qs
```

```python
words_qs = qs['words']
words_set_qs = set(" ".join(words_qs.tolist()).split())
len(words_set_qs)
```

```python
words_set_wv = set(wv.keys())
len(words_set_wv)
```

```python
len(words_set_wv & words_set_qs)
```

```python
# 使用wv生成embeding matrix
wem = np.zeros([WORDS_NUM + 1, WORD_EMBEDDING_DIM], dtype=np.float32)
wk2i = {}
for i, (k, v) in enumerate(wv.items()):
    wem[i + 1] = v
    wk2i[k] = i + 1
```

```python
cv = pickle.load(open(CHAR_EMBED_PKL, 'rb'))
chars_set_cv = set(cv.keys())
len(cv), len(set(cv.keys()))
```

```python
chars_qs = qs["chars"]
chars_set_qs = set(" ".join(chars_qs.tolist()).split())
len(chars_set_qs), len(set(cv.keys()) & chars_set_qs)
```

```python
len(cv[list(cv.keys())[0]])
```

```python
# 使用cv生成embeding matrix
cem = np.zeros([CHARS_NUM + 1, CHAR_EMBEDDING_DIM], dtype=np.float32)
ck2i = {}
for i, (k, v) in enumerate(cv.items()):
    cem[i + 1] = v
    ck2i[k] = i + 1
```

```python
# 保存wem，cem，wk2i，ck2i
WEM_PKL = os.path.join(INTERMEDIATE_DATA_PATH, 'WEM.pkl')
CEM_PKL = os.path.join(INTERMEDIATE_DATA_PATH, 'CEM.pkl')
with open(WEM_PKL, 'wb') as f:
    pickle.dump({'embedding_matrix': wem, 'word2index':wk2i}, f)

with open(CEM_PKL, 'wb') as f:
    pickle.dump({'embedding_matrix':cem, 'char2index':ck2i}, f)
```
